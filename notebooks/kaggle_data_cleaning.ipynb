2 â€” Carga del archivo

```python
# =============================
# CARGA DEL DATASET
# =============================


print("ğŸ“‚ Seleccionamos 'raw_kaggle_data.csv'")

# Almacenando en variable la ruta
CSV_NAME = "../src/preprocessing/data/raw/raw_kaggle_data.csv"
#leer
df = pd.read_csv(CSV_NAME)
print("âœ… Dataset cargado con Ã©xito")
print(f"Filas: {df.shape[0]}, Columnas: {df.shape[1]}")
df.head()
```

4 â€” Reemplazo de sÃ­mbolos extraÃ±os

```python
# =============================
# LIMPIEZA: SÃMBOLOS EXTRAÃ‘OS â†’ NaN
# =============================
print("\nğŸ§¹ Reemplazando sÃ­mbolos raros ('?', '###') por NaN...")
df = df.replace(["?", "###"], np.nan)

print("\nğŸ” Nuevos conteos de valores nulos tras reemplazo:")
print(df.isnull().sum())
```

5 â€” Regla de negocio y imputaciÃ³n

```python
# =============================
# REGLA DE NEGOCIO + IMPUTACIÃ“N
# =============================

# Copia para comparaciones al final
df_original = df.copy()

# OpciÃ³n A: eliminar filas con > umbral de NaN
umbral = 0.3  # si mÃ¡s del 30% de columnas estÃ¡n vacÃ­as -> eliminar fila
df_cleaned = df.dropna(thresh=int(df.shape[1] * (1 - umbral))).copy()

# OpciÃ³n B: imputar
for col in df_cleaned.columns:
    if df_cleaned[col].dtype == 'object' or pd.api.types.is_bool_dtype(df_cleaned[col]):
        # modo puede fallar si todo es NaN; controlamos eso
        if df_cleaned[col].dropna().empty:
            df_cleaned[col] = df_cleaned[col].fillna("Desconocido")
        else:
            df_cleaned[col] = df_cleaned[col].fillna(df_cleaned[col].mode()[0])
    else:
        df_cleaned[col] = pd.to_numeric(df_cleaned[col], errors="coerce")
        df_cleaned[col] = df_cleaned[col].fillna(df_cleaned[col].mean())

print("\nâœ… Limpieza completada. Info del dataset limpio:")
print(df_cleaned.info())
```

6 â€” SelecciÃ³n de variables relevantes

```python
# =============================
# SELECCIÃ“N DE VARIABLES
# =============================
variables_mantener = [
    # DesempeÃ±o acadÃ©mico - 1er semestre
    'Curricular units 1st sem (enrolled)',
    'Curricular units 1st sem (evaluations)',
    'Curricular units 1st sem (approved)',
    'Curricular units 1st sem (grade)',

    # DesempeÃ±o acadÃ©mico - 2do semestre
    'Curricular units 2nd sem (enrolled)',
    'Curricular units 2nd sem (evaluations)',
    'Curricular units 2nd sem (approved)',
    'Curricular units 2nd sem (grade)',

    # InformaciÃ³n personal/demogrÃ¡fica
    'Age at enrollment',
    'Scholarship holder',
    'Gender',
    "Mother's qualification",
    "Father's qualification",
    "Mother's occupation",
    "Father's occupation",
    'Displaced',
    'Educational special needs',

    # SituaciÃ³n econÃ³mica/acadÃ©mica
    'Tuition fees up to date',
    'Debtor',
    'Daytime/evening attendance',
    'Previous qualification',

    # Variable objetivo
    'Target'
]

print(f"\nğŸ“Š ANTES de la selecciÃ³n: {df_cleaned.shape[1]} columnas")

variables_existentes = [c for c in variables_mantener if c in df_cleaned.columns]
variables_no_encontradas = [c for c in variables_mantener if c not in df_cleaned.columns]

df_final = df_cleaned[variables_existentes].copy()

print(f"ğŸ“Š DESPUÃ‰S de la selecciÃ³n: {df_final.shape[1]} columnas")
print(f"âœ… Variables encontradas: {len(variables_existentes)}")
print(f"âŒ Variables no encontradas: {len(variables_no_encontradas)}")
if variables_no_encontradas:
    print("No se encontraron:")
    for v in variables_no_encontradas:
        print("  -", v)
```

7 â€” ConversiÃ³n de tipos

```python
# =============================
# CONVERSIÃ“N DE TIPOS
# =============================
conversion_tipos = {
    'Age at enrollment': 'int64',
    'Scholarship holder': 'int64',
    'Gender': 'int64',
    "Mother's qualification": 'int64',
    "Father's qualification": 'int64',
    "Mother's occupation": 'int64',
    "Father's occupation": 'int64',
    'Displaced': 'int64',
    'Educational special needs': 'int64',
    'Tuition fees up to date': 'int64',
    'Debtor': 'int64',
    'Daytime/evening attendance': 'int64',
    'Previous qualification': 'int64',
    'Curricular units 1st sem (enrolled)': 'int64',
    'Curricular units 1st sem (evaluations)': 'int64',
    'Curricular units 1st sem (approved)': 'int64',
    'Curricular units 1st sem (grade)': 'float64',
    'Curricular units 2nd sem (enrolled)': 'int64',
    'Curricular units 2nd sem (evaluations)': 'int64',
    'Curricular units 2nd sem (approved)': 'int64',
    'Curricular units 2nd sem (grade)': 'float64',
    'Target': 'object'
}

print("\nğŸ“‹ Tipos ANTES:")
print(df_final.dtypes)

print("\nğŸ”§ Convirtiendo tipos de datos...")
for columna, tipo_deseado in conversion_tipos.items():
    if columna in df_final.columns:
        try:
            if tipo_deseado == 'int64':
                df_final[columna] = pd.to_numeric(df_final[columna], errors='coerce').astype('float64').astype('int64')
            elif tipo_deseado == 'float64':
                df_final[columna] = pd.to_numeric(df_final[columna], errors='coerce').astype('float64')
            else:
                df_final[columna] = df_final[columna].astype('object')
            print(f"âœ… {columna} â†’ {tipo_deseado}")
        except Exception as e:
            print(f"âŒ Error en {columna} ({tipo_deseado}): {e}")

print("\nğŸ“‹ Tipos DESPUÃ‰S:")
print(df_final.dtypes)

```
